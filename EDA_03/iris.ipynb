Данные содержат 150 экземпляров ириса, по 50 экземпляров трех видов — Ирис щетинистый (Iris setosa), 
Ирис виргинский (Iris virginica) и Ирис разноцветный (Iris versicolor). 
Для каждого экземпляра измерялись четыре характеристики (в сантиметрах):


sepal length — длина наружной доли околоцветника;

sepal width — ширина наружной доли околоцветника;

petal length — длина внутренней доли околоцветника;

petal width — ширина внутренней доли околоцветника.

```python
import pandas as pd
import matplotlib.pyplot as plt # библиотека визуализации
import seaborn as sns
```

```python
iris = pd.read_csv('data/iris.zip')
iris.head()
```

```python
plt.figure(figsize=(14, 10))
sns.heatmap(iris.corr(), annot=True)
```

```python
# petal.length находится в парах двух самых высоких корреляций, начинаю удаление с него
iris.drop(['petal.length'], axis=1, inplace=True)
```

```python
# Опять смотрю на корреляции
plt.figure(figsize=(14, 10))
sns.heatmap(iris.corr(), annot=True)
```

```python
# petal.width находится в парах с самыми высокими корреляциями по модулю 0,82 и -0,37
iris.drop(['petal.width'], axis=1, inplace=True)
```

```python
# Опять смотрю на корреляции
plt.figure(figsize=(14, 10))
sns.heatmap(iris.corr(), annot=True)
```
