Создадим обучающий набор для кодирования порядковых признаков — ассортимент небольшого магазина с одеждой, 
где size — буквенное обозначение размера одежды, type — тип изделия.

```python
# инициализируем информацию об одежде
clothing_list = [
    ['xxs', 'dress'],
    ['xxs', 'skirt'],
    ['xs', 'dress'],
    ['s', 'skirt'],
    ['m', 'dress'],
    ['l', 'shirt'],
    ['s', 'coat'],
    ['m', 'coat'],
    ['xxl', 'shirt'],
    ['l', 'dress']
]

clothing = pd.DataFrame(clothing_list, columns = ['size',  'type'])
clothing
```

```python
import category_encoders as ce

# Нужен маплист для правильного, вручную, определения очередности элементов категории
maplist = [{
    'col': 'size', 
    'mapping': {
        'xxs': 0, 
        'xs': 1,
        's': 2, 
        'm': 3, 
        'l': 4, 
        'xxl': 5
        }
    }]

oe = ce.OrdinalEncoder(mapping=maplist)
clothing = oe.fit_transform(clothing, return_df=True)
```

```python
clothing
```

```python
encoder = ce.OneHotEncoder(cols=['type'], use_cat_names=True) # указываем столбец для кодирования
clothing_hot = encoder.fit_transform(clothing)

clothing_hot
```

```python
bin_encoder = ce.BinaryEncoder(cols=['type']) # указываю столбцы для кодирования
clothing_bin = bin_encoder.fit_transform(clothing)

clothing_bin
```
