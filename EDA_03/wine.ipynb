Совсем ненормальное распределение. Необходимо использовать RobustScaler

```python
data = pd.read_csv('data\wine_cleared.zip', index_col=0)
```

```python
data['price'].describe()
```

```python
data.info()
```

```python
# для нормализации, стандартизации
from sklearn import preprocessing

# инициализируем нормализатор RobustScaler
r_scaler = preprocessing.RobustScaler()

# нормализую цену
price_r = r_scaler.fit_transform(data[['price']])
```

```python
data_r = pd.DataFrame(price_r, columns=['price'])
```

```python
data_r
```

```python
round(data_r['price'].sum(), 0)
```

Стандартизируйте исходный признак price.

```python
s_scaler = preprocessing.StandardScaler()
scaler_price = s_scaler.fit_transform(data[['price']])
```
