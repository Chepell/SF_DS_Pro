```python
def f(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2
```

Градиент функции Розен-Брока. Функция по сути не используется дальше. Нужны частные производные.

```python
def grad(x, y):
    dx = 2 * x - 2 + 100 * (-4 * y * x + 4 * x**3)
    dy = 100 * (2 * y - 2 * x**2)
    return (dx, dy)
```

Задание 8.1

Найдите минимум функции с помощью градиентного спуска

```python
from sympy import *
```

```python
x, y = symbols('x y')
f = 2 * x**2 - 4 * x * y + y**4 + 2
f
```

Рассчитаю первые производные функции и найду градиент

```python
f_diff_x = f.diff(x)
f_diff_y = f.diff(y)
```

```python
f_diff_x
```

```python
f_diff_y
```

```python
def grad(x0, y0):
    dx = f_diff_x.subs({x: x0, y: y0})
    dy = f_diff_y.subs({x: x0, y: y0})
    return (dx, dy)
```

Подбираю начальную точку

```python
start_point = (-5, 8)
```

Градиент не должны быть равен нулю для начальной точки, иначе это уже минимум и отсюда такой простой алгоритм уже не выберется

```python
grad(*start_point)
```

```python
x0 = start_point # Начальная точка для запуска алгоритма
gamma = 0.001 # Темп обучения изначально 0.1 и очень большой, сразу выпрыгиваю из впадины и уже туда не возвращаюсь
x_cur = x0 # Сохраняю текущую координату

vals = []
coords = []

i = 0
while True:
    # получение новой точки на основе предыдущей и градиента, с минусом т.к. минимизируем функцию и используем антиградиент
    x_new = (x_cur[0] - gamma * grad(*x_cur)[0],
            x_cur[1] - gamma * grad(*x_cur)[1])
    
    # Обучение прерывается только по количеству итераций
    if i > 50000:
        break
    
    # сохраняю последнее найденное значение точки в переменную и добавляю в список
    x_cur = x_new
    coords.append(x_cur)
    # По точке считаю значение функции и сохраняю в список
    f_val = f.subs({x: x_cur[0], y: x_cur[1]})
    vals.append(f_val)
    
    i += 1 # обновляю счетчик
    
    # Вывожу данные на каждой итерации
    print(f"iter={i}; x=({x_cur[0]:.4f}, {x_cur[1]:.4f});"
          f" f(x)={f_val:.4f}; grad f(x)=({grad(*x_cur)[0]:.4f}, {grad(*x_cur)[1]:.4f})")
```

Задание 8.2

Найдите минимум функции с помошью градиентного спуска

```python
x, y, z = symbols('x y z')
f = x**3 - 2 * x**2 + y**2 + z**2 - 2 * x * y + x * z - y * z + 3 * z
f
```

```python
f_diff_x = f.diff(x)
f_diff_y = f.diff(y)
f_diff_z = f.diff(z)
```

```python
f_diff_x
```

```python
f_diff_y
```

```python
f_diff_z
```

```python
# Функция для расчета градиента фунции в заданной точке
def grad(x0, y0, z0):
    # Используются рассчитанные ранее значения частных произодных
    dx = f_diff_x.subs({x: x0, y: y0, z: z0})
    dy = f_diff_y.subs({x: x0, y: y0, z: z0})
    dz = f_diff_z.subs({x: x0, y: y0, z: z0})
    return (dx, dy, dz)
```

Подбираю начальную точку

```python
start_point = (0, 0, 0)
```

Градиент не должны быть равен нулю для начальной точки, иначе это уже минимум и отсюда такой простой алгоритм уже не выберется

```python
grad(*start_point)
```

```python
x0 = start_point # Начальная точка для запуска алгоритма
gamma = 0.001 # Темп обучения изначально 0.1 и очень большой, сразу выпрыгиваю из впадины и уже туда не возвращаюсь
x_cur = x0 # Сохраняю текущую координату

vals = []
coords = []

i = 0
while True:
    # получение новой точки на основе предыдущей и градиента, с минусом т.к. минимизируем функцию и используем антиградиент
    x_new = (x_cur[0] - gamma * grad(*x_cur)[0],
            x_cur[1] - gamma * grad(*x_cur)[1],
            x_cur[2] - gamma * grad(*x_cur)[2])
    
    # Обучение прерывается только по количеству итераций
    if i > 10000:
        break
    
    # сохраняю последнее найденное значение точки в переменную и добавляю в список
    x_cur = x_new
    coords.append(x_cur)
    # По точке считаю значение функции и сохраняю в список
    f_val = f.subs({x: x_cur[0], y: x_cur[1], z: x_cur[2]})
    vals.append(f_val)
    
    i += 1 # обновляю счетчик
    
    # Вывожу данные на каждой итерации
    print(f"iter={i}; x=({x_cur[0]:.4f}, {x_cur[1]:.4f}, {x_cur[2]:.4f});"
          f" f(x)={f_val:.4f}; grad f(x)=({grad(*x_cur)[0]:.4f}, {grad(*x_cur)[1]:.4f}, {grad(*x_cur)[2]:.4f})")
```

Покрутив точку старта, темп обучения и количество итераций пришел вот к такому результату.

Когда видно что вектор градиента приближается к нулю, а значение функции на уровне -7 практически не меняется.
