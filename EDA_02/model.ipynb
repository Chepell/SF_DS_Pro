```python
import pandas as pd
import matplotlib.pyplot as plt # библиотека визуализации
import seaborn as sns
```

```python
df = pd.read_csv('data\model.zip')
```

Набор данных содержит ~600 записей о девушках и семь признаков:

BMI — индекс массы тела (ИМТ)

year — год размещения модели в журнале

waist — обхват талии модели

hips — обхват бёдер модели

height — рост модели

weight — вес модели

waist/hip — соотношение обхвата талии и бёдер

```python
corr_data = df.corr()
corr_data
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr_data)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr_data, annot=True)
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=df, x="Waist/Hip", y="Waist")
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=df, x="Weight", y="Year")
```

Сравнив эти два графика, мы можем сказать, 
что в первой паре признаков связь видна более отчётливо,
точки больше прижаты друг к другу и выстроены визуально в линию. 
Эта линия похожа на рассматриваемую нами в прошлом юните линейную связь. 
Во второй паре признаков они распределены по всему пространству — 
трудно выявить основную массу признаков, связь кажется очень слабой.

```python
sns.pairplot(df)
```
