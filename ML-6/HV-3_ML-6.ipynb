```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
data = pd.read_excel('data/data_ford_price.xlsx', engine='openpyxl')
```

```python
X, y = data.drop(columns='price'), data['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=40)
```

```python
# Словарь с методом отбора признаков и списком отобранных признаков
selection_map = {
    'RME': ['year', 'cylinders', 'lat'],
    'SelectKBest': ['year', 'cylinders', 'odometer']
}
```

```python
for name, features in selection_map.items():
    X_new = X[features] # оставляю только отобранные с помощью данного метода признаки
    
    X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.3, random_state=40)
    
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    mae_train = mean_absolute_error(y_train, y_train_pred)
    mae_test = mean_absolute_error(y_test, y_test_pred)
    
    print(f'{name}. train MAE: {mae_train:.2f}')
    print(f'{name}. test MAE: {mae_test:.2f}')
```

Отбор с помощью SelectKBest демонстирует меньшее значение заданной метрики MAE, чем метода RME.

Использую метод SelectKBest для отбора признаков.
