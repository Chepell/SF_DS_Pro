```python
import pandas as pd
```

```python
ratings_movies = pd.read_csv('data/ratings_movies.csv', index_col=False)
```

```python
ratings_movies_df = ratings_movies.copy()
```

```python
ratings_movies_df.drop(columns='Unnamed: 0', inplace=True)
```

```python
ratings_movies_df
```

```python
#библиотека для регулярных выражений
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
# Создайте в таблице новый признак year_release, который соответствует году выпуска фильма.
# У скольких фильмов не указан год их выпуска?
ratings_movies_df['year_release'] = ratings_movies_df['title'].apply(get_year_release)
```

```python
# Добавьте в таблицу новый признак year_rating — год выставления оценки. 
# Создайте сводную таблицу, которая иллюстрирует зависимость среднего 
# рейтинга фильма от года выставления оценки и жанра.
ratings_movies_df['date'] = pd.to_datetime(ratings_movies_df['date'])
ratings_movies_df['year_rating'] = ratings_movies_df['date'].dt.year
```

```python
pt = ratings_movies_df.pivot_table(values='rating', index='year_release', columns='genres', aggfunc='mean')
pt
```

```python
# За весь период (с 1996 по 2018 год) сочетание жанров Action|Adventure ни разу не получало среднюю оценку ниже 3.
pt.loc[1996:, 'Action|Adventure'].dropna().min() > 3
```

```python
# Наилучшую оценку жанр Action|Adventure|Animation|Children|Comedy|IMAX получил в 2010 году.
pt.loc[:,'Action|Adventure|Animation|Children|Comedy|IMAX'].dropna()
# False
```

```python
# Среди сочетаний жанров, получивших наивысшую среднюю оценку в 2018 году, есть сочетание Animation|Children|Mystery.
filter_pt = pt.loc[2018,].dropna()
filter_pt[filter_pt == filter_pt.max()]
# False
```
