```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Создание экземпляра KMeans
kmeans = KMeans(n_clusters=3, random_state=42)

# Подгонка модели к данным и выполнение предсказаний
y_kmeans = kmeans.fit_predict(df)

# Применение PCA для уменьшения размерности до 2D для визуализации
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df)

# Создание DataFrame с результатами PCA и метками кластеров
principalDf = pd.DataFrame(data=principalComponents, columns=['Principal Component 1', 'Principal Component 2'])
principalDf['Cluster'] = y_kmeans

# Визуализация результатов кластеризации
plt.scatter(principalDf['Principal Component 1'], principalDf['Principal Component 2'], c=principalDf['Cluster'], cmap='viridis')
plt.show()
```
