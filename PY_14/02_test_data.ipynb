```python
import pandas as pd
```

```python
test_data = pd.read_csv('data/test_data.csv')
test_data
```

```python
# Ваша задача очистить данную таблицу от пропусков следующим образом:
# - Если признак имеет больше 50% пропущенных значений - удалите его
# - Для оставшихся данных: если в строке более 2 пропусков - удалите строку
# - Для оставшихся данных: числовые признаки заполните средним значением, а категориальные - модой
# Результат должен находиться в переменной df
```

```python
col_null_percent = test_data.isnull().mean() * 100
col_with_null = col_null_percent[col_null_percent > 0]
col_with_null
```

```python
clear_data = test_data.copy()
```

```python
# Порог в 50% минимального количества значений в столбце
thresh = clear_data.shape[0] * 0.5
```

```python
# Если признак имеет больше 50% пропущенных значений - удалите его
clear_data.dropna(how='any', thresh=thresh, axis=1, inplace=True)
```

```python
clear_data
```

```python
# Для оставшихся данных: если в строке более 2 пропусков - удалите строку
in_row = clear_data.shape[1]
clear_data.dropna(how='any', thresh=in_row-2, axis=0, inplace=True)
```

```python
clear_data
```

```python
# Для оставшихся данных: числовые признаки заполните средним значением, а категориальные - модой
values = {
    'one': clear_data['one'].mean(),
    'two': clear_data['two'].mean(),
    'four': clear_data['four'].mode()[0]
    }

clear_data.fillna(value=values, inplace=True)
```
