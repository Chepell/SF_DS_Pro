```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Numbers of time steps
T = 1000

# Initial price
P0 = 10

# Drift
mu = 0.001
```

```python
# last log price
last_p = np.log(P0)
last_p
```

```python
log_returns = np.zeros(T)
prices = np.zeros(T)

for t in range(T):
    # sample a log return
    r = 0.01 * np.random.randn()
    
    # compute the new log price
    p = last_p + r + mu
    
    # store the return and price
    log_returns[t] = r + mu
    prices[t] = np.exp(p) # from log to base value
    
    last_p = p # update last price
```

```python
plt.figure(figsize=(20, 8))
plt.plot(prices)
plt.title('Simulated prices');
```
