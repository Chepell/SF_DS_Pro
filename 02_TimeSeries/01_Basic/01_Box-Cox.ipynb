# Примеры обработки данных разными способами

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import boxcox
```

```python
# !wget -nc https://lazyprogrammer.me/course_files/airline_passengers.csv
```

```python
df = pd.read_csv('airline_passengers.csv', index_col='Month', parse_dates=True)
df
```

```python
df
```

### Базовая визуализация

```python
df['Passengers'].plot(figsize=(20, 12));
```

### Power

```python
df['Sqrt_Passengers'] = np.sqrt(df['Passengers'])
```

```python
df['Sqrt_Passengers'].plot(figsize=(20, 12));
```

### Log

```python
df['Log_Passengers'] = np.log(df['Passengers'])
```

```python
df['Log_Passengers'].plot(figsize=(20, 12));
```

### Box-Cox

```python
df['BoxCox_Passengers'], lam = boxcox(df['Passengers']) # Возвращает трансформированные данные и значение лямбда
```

```python
lam
```

```python
df['BoxCox_Passengers'].plot(figsize=(20, 12));
```

### Distributions

```python
df['Passengers'].hist(bins=20, figsize=(14, 8));
```

```python
df['Sqrt_Passengers'].hist(bins=20, figsize=(14, 8));
```

```python
df['Log_Passengers'].hist(bins=20, figsize=(14, 8));
```

```python
df['BoxCox_Passengers'].hist(bins=20, figsize=(14, 8));
```

Распределение после трансформации BoxCox наиболее отцентрированно, хвосты одинаковые
