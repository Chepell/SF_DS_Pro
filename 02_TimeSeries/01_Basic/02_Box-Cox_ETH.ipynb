# Примеры обработки данных разными способами

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import boxcox
```

```python
df = pd.read_parquet('..\..\ETH-Full-1H.parquet') # 2 dir up
```

```python
df
```

### Базовая визуализация

```python
df['Close'].plot(figsize=(20, 12));
```

### Power

```python
df['Sqrt_Close'] = np.sqrt(df['Close'])
```

```python
df['Sqrt_Close'].plot(figsize=(20, 12));
```

### Log

```python
df['Log_Close'] = np.log(df['Close'])
```

```python
df['Log_Close'].plot(figsize=(20, 12));
```

### Box-Cox

```python
df['BoxCox_Close'], lam = boxcox(df['Close']) # Возвращает трансформированные данные и значение лямбда
```

```python
lam
```

```python
df['BoxCox_Close'].plot(figsize=(20, 12));
```

### Distributions

```python
df['Close'].hist(bins=20, figsize=(14, 8));
```

```python
df['Sqrt_Close'].hist(bins=20, figsize=(14, 8));
```

```python
df['Log_Close'].hist(bins=20, figsize=(14, 8));
```

```python
df['BoxCox_Close'].hist(bins=20, figsize=(14, 8));
```

Распределение после трансформации BoxCox наиболее отцентрированно, хвосты одинаковые
